# Example of a line plot

```python
import matplotlib.pyplot as plt                #importing matplotlib as a function of pyplot
import numpy as np                             #import numphy array

xpoints = np.array([1, 2, 6, 8])               #specify x points (set 1)
ypoints = np.array([3, 8, 1, 10])              #Specify y points (set 1)
x2 = np.array([1, 3, 7, 10])                   #Specify x points (set 2)
y2 = np.array([2, 6, 1, 9])                    #Specify y points (set 2)

plt.subplot(1, 2, 1) # this creates a subplot, the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(xpoints, ypoints,'o:y',linewidth = '4',ms = 5, mfc = 'b', mec = 'r')  #plot set 1
plt.title("Graph 1",fontdict = font1)
plt.subplot(1, 2, 2) # this creates a subplot, #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(x2, y2,'o:r',linewidth = '4',ms = 5, mfc = 'r', mec = 'g')            #plot set 2
plt.title("Graph 2",fontdict = font1)

## note for subplots to put them on top of each other use (2,1,1) and (2,1,2)


##Note the following:ms (sets size of points), mfc (sets color of points), mec (sets color of line), linewidth (sets thickness of line)

font1 = {'family':'serif','color':'blue','size':10}       # These parameters set font style and size for titles and axis
font2 = {'family':'serif','color':'darkred','size':15}

plt.suptitle("Matplotlib Graph",fontdict = font1)     #fontdict assigns specified font styles to heading 
plt.xlabel("Test Label X",fontdict = font2)
plt.ylabel("Test Label Y",fontdict = font2)

# to create a scatter replace above with plt.scatter

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)   # adds grid to the graph

plt.show()    # displays/plots graph
```

#### The following is an example of a bar graph being used to illustrate cycle time reduction in decorative plant

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Q1", "Q2", "Q3", "Q4"])
Manufacturing = np.array([21.5, 36.4, 30.0, 28.9])
Filling = np.array([10.5, 13.0, 12.3, 12.5])

# benchmark line graph
yline = np.array([33,33,33,33,33])
plt.plot(yline,'o:g',linewidth = '2')  ##color changes color of point

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)   # adds grid to the graph

plt.suptitle("Stacked Chart Cycle Time Results 2019",fontdict = font1)     #fontdict assigns specified font styles to heading 
plt.xlabel("Quarter",fontdict = font2)
plt.ylabel("Average Cycle Time(hrs)",fontdict = font2)

plt.bar(x,Filling, color='r')
plt.bar(x,Manufacturing,bottom=Filling, color='b')
plt.show()
```

# Examples of Distribution Graphs

## Normal Distribution

```python
from numpy import random          ##numpy array requried to graph data
import matplotlib.pyplot as plt   ## function of matplotlib
import seaborn as sns  ## seaborn is used to create distributions


sns.displot(random.normal(size=1000))   # plot normal distribution
sns.displot(random.binomial(n=100, p=0.5, size=1000), label='binomial')  # plot binomial distribution
sns.displot(random.poisson(lam=2, size=1000), label='poisson') #plot poisson distribution
sns.displot(random.uniform(size=1000)) #uniform distribution
sns.displot(random.logistic(size=1000)) #logistic distribution
sns.displot(random.exponential(size=1000)) #exponential distribution
sns.displot(random.chisquare(df=2, size=1000)) #chi sauare distribution
sns.displot(random.rayleigh(size=1000)) #rayleigh distribution
sns.displot(random.pareto(a=2, size=1000), kde=False)
x = random.zipf(a=2, size=1000)
sns.displot(x[x<10], kde=False)

plt.show()
```
